```python
import os
import xarray as xr
```

```python
# Use the current working directory but go one directory up to the project root
base_dir = os.path.dirname(os.getcwd())  # Go up one directory level

# Define the status quo and optimized directories
statusquo_dir = os.path.join(base_dir, 'data', 'statusquo')
#optimized_dir = os.path.join(base_dir, 'data', 'opti')

# Relative paths to your NetCDF files
statusquo_file_light = os.path.join(statusquo_dir, 'Playground_2024-07-06_04.00.00_light.nc')
#optimized_file = os.path.join(optimized_dir, 'Playground_2024-07-06_04.00.00.nc')
```

```python
# Load the light version of the dataset
#file_path = 'data/statusquo/Playground_light.nc'  # Adjust path if needed
file_path = statusquo_file_light
ds_light = xr.open_dataset(file_path)

# View the dataset structure
print(ds_light)

# Inspect the variables
print("Variables in the dataset:")
print(ds_light.data_vars)

# Get the time dimension to confirm it exists
if 'Time' in ds_light:
    print(f"Time dimension: {ds_light['Time'].values}")

# Display a specific variable (adjust the variable name as per your needs)
if 'TSurf' in ds_light:
    print("Surface temperature data (TSurf):")
    print(ds_light['TSurf'].values)

# Optionally, plot a variable
ds_light['TSurf'].isel(Time=0).plot(cmap='coolwarm')  # Plot for first time step, adjust if necessary
```
